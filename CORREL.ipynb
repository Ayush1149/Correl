```python
 #Defining the imaginary data in Python as lists
companies = ["Alpha Corp", "Beta Industries", "Gamma Solutions", "Delta Enterprises", "Epsilon Tech", 
             "Zeta Services", "Eta Holdings", "Theta Systems", "Iota Corp", "Kappa Group"]
```

```python
# Advertising Budget (in million USD)
advertising_budget = [2.5, 3.8, 1.7, 4.2, 2.9, 3.0, 3.7, 2.6, 3.5, 2.0]
```

```python
# Sales Revenue (in million USD)
sales_revenue = [45.3, 58.7, 39.1, 62.4, 48.6, 53.2, 60.0, 47.5, 55.7, 41.8]
```

```python
data = list(zip(companies, advertising_budget, sales_revenue))
data
```

```python
# Importing necessary libraries for plotting
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(advertising_budget, sales_revenue, color='green', label='Companies')

# Add labels for each point
for i, company in enumerate(companies):
    plt.text(advertising_budget[i] + 0.05, sales_revenue[i] - 0.5, company, fontsize=9)# use this code after plotting the whole graph

# Best fit line (linear regression)
m, b = np.polyfit(advertising_budget, sales_revenue, 1)
plt.plot(advertising_budget, m*np.array(advertising_budget) + b, color='red', label=f'Best Fit Line (y={m:.2f}x + {b:.2f})')

# Graph details
plt.title("Correlation Between Advertising Budget and Sales Revenue")
plt.xlabel("Advertising Budget (in million USD)")
plt.ylabel("Sales Revenue (in million USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
